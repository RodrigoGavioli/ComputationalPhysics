```python

Valores = np.linspace(0,2*np.pi,1000)
fx = np.zeros(1000)
fy = np.zeros(1000)
i = 0
for theta in Valores:
    
    fx[i] = 2*np.cos(theta) + np.cos(2*theta)
    fy[i] = 2*np.sin(theta) - np.sin(2*theta)
    i+= 1
plt.plot(fx,fy)
plt.show()
    
```

```python
#2 
i = 0
gX = np.zeros(1000)
gY = np.zeros(1000)
def EG(theta):
    return theta**2
values = np.linspace(0,10*np.pi,1000)
R = np.zeros(1000)
for v in values:
    R[i] = EG(v)
    gX[i] = EG(v)*np.cos(v)
    gY[i] = EG(v)*np.sin(v)
    i += 1
plt.plot(gX,gY)
plt.show()
    

```

```python
def Fey(theta):
    return (np.exp(np.cos(theta))-2*np.cos(4*theta)+(np.sin(theta/12))**5)

hX = np.zeros(1000)
hY = np.zeros(1000)
i = 0
ptheta = np.linspace(0,24*np.pi,1000)
for t in ptheta:
    hX[i] = Fey(t)*np.cos(t)
    hY[i] = Fey(t)*np.sin(t)
    i += 1
plt.plot(hX,hY)
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

Z = np.loadtxt("stm.txt")
print(Z)

Ny, Nx = Z.shape
print(Ny)
x = np.arange(Nx)
y = np.arange(Ny)
X, Y = np.meshgrid(x, y, indexing="xy")

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("altura")
plt.show()
```
