```python
def Catalan():
    LNumber = []
    C = 1
    n = 0
    while C < 10**9:
        n += 1
        LNumber.append(C)
        C *= ((4*n + 2)/(n+2))
    return LNumber

print(Catalan())
```

```python
def CatalanRecursivo(n):
    if n == 0: 
        return 1
    else:
        return ((4*n-2)/(n+1))*CatalanRecursivo(n-1)
    


print(CatalanRecursivo(100))
    
```

```python
def factorial(n):
    n = int(n)
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(200.))
```

```python

def Potencial(L):
    L = int(L)
    soma = 0
    for i in range(-L,L):
        for j in range(-L,L):
            for k in range(-L,L):
                
                soma += 1/np.sqrt(i**2+j**2+k**2)
    return soma


def potencial(i,j,k,soma):
    if i == 0 and j == 0 and k == 0:
        return 0
    pot = 1/np.sqrt(i**2+j**2+k**2)
    if (i+j+k) %2 != 0:
        pot = -pot
    
    return soma + pot + potencial(i+1,j,k,soma) + potencial(i,j+1,k,soma) + potencial(i,j,k+1,soma)
    
    
def Calculo(L):
    return potencial(-L,-L,-L,0)

print(Calculo(10))




```
