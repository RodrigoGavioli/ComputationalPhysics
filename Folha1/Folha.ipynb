```python
#1)
def factorial(i):
    if i == 0:
        return 1
    else: 
        return i*factorial(i-1)
    
def binomial(n,k):
    if k == 0:
        return 1
    else:
        return factorial(n)/(factorial(k)*factorial(n-k))
    

print(binomial(20,1))

```

```python
#2)
def triangulopascal(n):
    linha = 0
    while linha <= n:
        for i in range(linha):
            print(binomial(linha,i))
        linha += 1
    return(f"este acima é o triangulo de pascal com altura: {n}")
print(triangulopascal(20))
```

```python
#3)
def Probabilidade(n,k):
    return (1/(2**n))*(binomial(n,k))

print(Probabilidade(100,60))#prob de obter 60 caras lanç 100vezes
soma = 0
for i in range(60,101):
    soma += Probabilidade(100,i)

print(soma) #prob de obter pelo menos 60 caras em 100 lanç
```
