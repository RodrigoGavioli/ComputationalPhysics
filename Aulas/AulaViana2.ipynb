```python
#vimos eliminação de gauss e gaussjordan
#sendo a elim. de gauss é melhor
#elim. de gauss aplicada à uma matriz para construir a inversa 
from numpy import array,empty

A = array([[ 4,  -1,  -1,  -1 ],
           [ -1,  3, 0, -1 ],
           [ -1, 0, 3,  -1 ],
           [ -1, -1,  -1,  4 ]],float)
v = array([ 5, 0, 5, 0 ],float)
N = len(v)

# Gaussian elimination
for m in range(N):

    # Divide by the diagonal element
    div = A[m,m]
    A[m,:] /= div
    v[m] /= div

    # Now subtract from the lower rows
    for i in range(m+1,N):
        mult = A[i,m]
        A[i,:] -= mult*A[m,:]
        v[i] -= mult*v[m]

# Backsubstitution
x = empty(N,float)
for m in range(N-1,-1,-1):
    x[m] = v[m]
    for i in range(m+1,N):
        x[m] -= A[m,i]*x[i]

print(x)


```

```python
import numpy as np

A = np.array([[ 4,  -1,  -1,  -1 ],
              [ -1,  0,  0,  -1 ],
              [ -1,  0,  3,  -1 ],
              [ -1, -1, -1,  4 ]], float)
v = np.array([5, 0, 5, 0], float)
N = len(v)

# Guardar a ordem das variáveis (já que vamos trocar colunas)
ordem = np.arange(N)

for m in range(N):
    # Procurar o maior elemento da submatriz A[m:,m:]
    maior = 0
    linha = m
    coluna = m
    for i in range(m, N):
        for j in range(m, N):
            if abs(A[i,j]) > maior:
                maior = abs(A[i,j])
                linha = i
                coluna = j
    
    # Trocar linhas
    A[[m, linha]] = A[[linha, m]]
    v[[m, linha]] = v[[linha, m]]
    
    # Trocar colunas (e registrar que trocamos)
    A[:, [m, coluna]] = A[:, [coluna, m]]
    ordem[[m, coluna]] = ordem[[coluna, m]]
    
    # Dividir a linha m pelo pivô
    div = A[m,m]
    A[m,:] /= div
    v[m] /= div
    
    # Eliminação
    for i in range(m+1, N):
        mult = A[i,m]
        A[i,:] -= mult*A[m,:]
        v[i] -= mult*v[m]

print("Matriz A escalonada:")
print(A)
print("Vetor v escalonado:")
print(v)
print("Ordem das variáveis após pivotagem:")
print(ordem)
```

```python
def eliminacao_gauss(A, b):
    n = len(A)
    for i in range(n):
        for j in range(i+1, n):
            fator = A[j,i] / A[i,i]
            A[j,i:] -= fator * A[i,i:]
            b[j] -= fator * b[i]
    x = np.zeros(n)
    for i in reversed(range(n)):
        x[i] = (b[i] - np.dot(A[i,i+1:], x[i+1:])) / A[i,i]
    return x

# Exemplo
A = np.array([[2,1,-1],[-3,-1,2],[-2,1,2]], dtype=float)
b = np.array([8,-11,-3], dtype=float)
print(eliminacao_gauss(A, b))
```

```python
def gauss_jordan(A, b):
    n = len(A)
    for i in range(n):
        A[i] = A[i] / A[i,i]
        b[i] = b[i] / A[i,i]
        for j in range(n):
            if j != i:
                fator = A[j,i]
                A[j] -= fator * A[i]
                b[j] -= fator * b[i]
    return b

# Exemplo
A = np.array([[2,1,-1],[-3,-1,2],[-2,1,2]], dtype=float)
b = np.array([8,-11,-3], dtype=float)
print(gauss_jordan(A, b))
```
