## Regra do trapezio

```python
def trapezio(a,b,N):
    h = (b-a)/N
    integral = 0
    for k in range(1,N):
        integral += f(a+k*h)
    integral += 1/2 * (f(a) + f(b))
    return h*integral 
def f(x):
    return x**2 
print(trapezio(0,1,100))

#Acho que esta correto Preciso confirmar com o Sr. GPT
```

## Regra de Simpson

```python
def simpson(a,b,N):
    h = (b-a)/N
    integral = f(a)+f(b)
    for kodd in range(1,N,2):
        integral += 4*(f(a + kodd*h))
    for keven in range(2,N,2):
        integral += 2*(f(a + keven*h))
    
    return (h/3)* integral 

print(simpson(0,1,6))
```

## Erros no integral de trapezio

```python
def errotrapezio(a, b, epsilon):
    N = 2
    h = (b-a)/N
    integral1 = (h/2)*(f(a)+f(b))
    erro = 1
    while erro > epsilon:
        N += 2
        h = (b-a)/N
        integral2 = (f(a)+f(b))/2
        for k in range(1,N):
            integral2 += f(a+k*h)
        integral2 *= h
        erro = abs(integral2 - integral1)
        integral1 = integral2
    return print(f"Conseguimos a integral: {integral1} com um erro de {erro} e com {N} divisoes")
print(errotrapezio(0,1,1e-7))
```

## Quadratura Gaussiana

```python
A = np.array([[1, 4, 8, 4],
              [4, 2, 3, 7],
              [8, 3, 6, 9],
              [4, 7, 9, 2]], float)
print(A[0,:])
A[1,:] -= A[0,:]*A[1,0]
print(A)
```

```python
A = np.array([[ 4,  -1,  -1,  -1 ],
              [ -1,  5,  0,  -1 ],
              [ -1,  0,  3,  -1 ],
              [ -1, -1, -1,  4 ]], float)
C = np.copy(A)
v = np.array([5,0,5,0])

def gaussian_elimination(A,v):
    C = np.copy(A)
    b = np.copy(v)
    N = len(C)
    for i in range(N):
        pivot = C[i,i]
        C[i,:] /= pivot
        b[i] /= pivot
        for j in range(i+1,N):
            fator = C[j,i]
            C[j,:] = C[j,:] - C[i,:]*fator
            b[j] = b[j]-b[i]*fator
    #Agora fazer o backsubstitution
    x = np.empty(N,float)
    for k in range(N-1,-1,-1):
        x[k] = b[k]
        for i in range(k+1,N):
            x[k] -= C[k,i]*x[i]
    
    return x

print(gaussian_elimination(A,v))
```

```python
A = np.array([[ 4,  -1,  -1,  -1 ],
              [ -1,  5,  0,  -1 ],
              [ -1,  0,  3,  -1 ],
              [ -1, -1, -1,  4 ]], float)

v = np.array([5, 0, 5, 0], float)
print(len(v))
def gaussian_elimination(A, v):
    A = A.copy()
    v = v.copy()
    N = len(A)

    for i in range(N):
        pivot = A[i, i]
        if pivot == 0:
            raise ValueError("Pivot zero encontrado — necessário pivotamento parcial.")
        
        for j in range(i+1, N):
            fator = A[j, i] / pivot
            A[j, :] -= fator * A[i, :]
            v[j] -= fator * v[i]
    return v

print(gaussian_elimination(A,v))
```
