```python
import numpy as np 
import matplotlib.pyplot as plt

def simpson(f,h):
    return (h/3)*(f[0] + f[-1] + 4*np.sum(f[1:-1:2])+ 2*np.sum(f[2:-2:2]))

def mettrap(f,h, N):
    return (h/2)*(f[0] + f[-1] + 2*np.sum(f[1:(N-1):1]))

def f(x):
    return x**2

N = 100
a = 0
b = 1
h = (b-a)/N
x = np.linspace(a, b ,N+1)
Y = f(x)
print(f"This is simpson:{simpson(Y,h)}")
print(f"This is mettrapezios{mettrap(Y,h,N)}")

```

```python
def romberg(f, a, b, N):
    R = np.zeros((N,N))
    R[0, 0] = (b-a)/2 *(f(a)+f(b))

    for i in range(1, N):
        h = (b-a)/(2**(i))
        R[i, 0] = R[i-1, 0]/2 + h*np.sum((f(a +(2*k-1)*h) for k in range(1,2**(i-1)+1))) 
        for j in range(1, i+1):
            R[i, j] = R[i, j-1] + (R[i, j-1]-R[i-1, j-1])/(4**(j)-1)
        
    return R
```

```python
import numpy as np 
import matplotlib.pyplot as plt

def simpson(f,h):
    return (h/3)*(f[0] + f[-1] + 4*np.sum(f[1:-1:2])+ 2*np.sum(f[2:-2:2]))



def f(x):
    return np.sqrt(x+1)
a = 0
b = 1
N = 2**0
Narray = np.array([])
kn = 0
vreal = (2/3)*(2**(1.5)-1)
Integrais = np.array([])
desviorelativo = np.array([])
for i in range(10):
    N = 2**i
    Narray = np.append(Narray, N)
    x = np.linspace(a,b,N+1)
    y = f(x)*np.exp(-1j*kn*x)
    h = (b-a)/N
    Integrais = np.append(Integrais, simpson(y,h))

desviorelativo = abs((Integrais - vreal)/vreal)


plt.plot(Narray, desviorelativo)
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
def f(x):
    return np.exp((-40*(1-(np.sin(np.pi*x)**2))))
N = 100
x = np.linspace(a, b, 100)
y = f(x)
n = 2

def coefkn(n):
    kn = []
    for i in range(-n,n+1):
        kn.append(2*np.pi*i/b)
    return kn

def coef(a, b, kn, n, N):
    h = (b-a)/N
    cn = []
    for i in range(len(kn)):
        y = f(x) * np.exp(-1j*kn[i]*x)
        cn.append(simpson(y,h))
    return cn

def somafourier(x, kn, cn):
    soma = np.zeros_like(x, dtype=complex)
    for i in range(len(kn)):
        soma += cn[i] * np.exp(1j * kn[i] * x)
    return soma

        
kn = coefkn(n)
cn = coef(a, b, kn, n, N)
fourier = somafourier(x, kn, cn)

plt.plot(x, fourier.real)
plt.plot(x, fourier.imag)
plt.show()
```
