```python
#Fazer o metodo da relaxação 
import numpy as np
import matplotlib.pyplot as plt

Ifinal = 2
Points = 1000
accuracy = 1e-6
v = 0.01
times = 0

def relaxation_method(v,c,f,max,accuracy,times):
    for i in range(max):
        times += 1
        v,oldv = f(v,c),v
        if abs(v-oldv) < accuracy:
            break
    return v, times
print(f"encontramos o valor {relaxation_method(v,2,f,Points,accuracy,times)[0]} em {relaxation_method(v,2,f,Points,accuracy,times)[1]} iterações")
```

```python
#Overrelaxatation
w = 1.
v = 0.01
c = 2
max = 1000
accuracy = 1e-6
times = 0
def overrelaxation_method(w,v,c,f,max,accuracy,times):
    for i in range(max):
        w = int(w)
        times += 1
        v,oldv = (1+w)*f(v,c)-w*v,v
        if abs(v-oldv) < accuracy:
            break
    return v, times

print(f"O metodo de overelaxation convergiu com um valor de: {overrelaxation_method(w,v,c,f,max,accuracy,times)[0]} em {overrelaxation_method(w,v,c,f,max,accuracy,times)[1]} tentativas ")
print(f"encontramos o valor {relaxation_method(v,2,f,Points,accuracy,times)[0]} em {relaxation_method(v,2,f,Points,accuracy,times)[1]} iterações")
```
