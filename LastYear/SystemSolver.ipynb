```python
#QR algorithm with eigenvalues

import numpy as np

def qr_decomposition_modified(A):
    N = len(A)
    Q = np.zeros((N, N))
    R = np.zeros((N, N))
    V = np.copy(A)

    for i in range(N):
        R[i,i] = np.linalg.norm(V[:,i])
        Q[:,i] = V[:,i] / R[i,i]
        for j in range(i+1, N):
            R[i,j] = np.dot(Q[:,i], V[:,j])
            V[:,j] = V[:,j] - R[i,j] * Q[:,i]

    return Q, R

def qr_algorithm(A, tol=1e-6, max_iterations=1000):
    A_k = np.copy(A)

    for iteration in range(max_iterations):
        Q, R = qr_decomposition_modified(A_k)
        A_k = R @ Q

        off_diagonal = A_k - np.diag(np.diag(A_k))
        if np.all(np.abs(off_diagonal) < tol):
            print(f"\nConvergiu em {iteration} iterações")
            break

    eigenvalues = np.diag(A_k)
    return eigenvalues, A_k

# Teste
A = np.array([[1, 4, 8, 4],
              [4, 2, 3, 7],
              [8, 3, 6, 9],
              [4, 7, 9, 2]], float)

eigenvalues, A_final = qr_algorithm(A)

print("\nAutovalores encontrados:")
print(eigenvalues)

print("\nMatriz final aproximada (quase diagonal):")
print(A_final)
```

```python
#Gaussian Quadrature 

def gaussxw(N):

    # Initial approximation to roots of the Legendre polynomial
    a = np.linspace(3,4*N-1,N)/(4*N+2)
    x = np.cos(np.pi*a+1/(8*N*N*np.tan(a)))

    # Find roots using Newton's method
    epsilon = 1e-15
    delta = 1.0
    while delta>epsilon:
        p0 = np.ones(N,float)
        p1 = np.copy(x)
        for k in range(1,N):
            p0,p1 = p1,((2*k+1)*x*p1-k*p0)/(k+1)
        dp = (N+1)*(p0-x*p1)/(1-x*x)
        dx = p1/dp
        x -= dx
        delta = max(abs(dx))

    # Calculate the weights
    w = 2*(N+1)*(N+1)/(N*N*(1-x*x)*dp*dp)

    return x,w

def gaussxwab(N,a,b):
    x,w = gaussxw(N)
    return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w
```
