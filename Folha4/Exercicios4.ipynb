Exercicio 4.1 e 4.2 

```python
N = 10
a = 0
b = 2 
Vreal = 4.4


def f(x):
    return x**4 - 2*x + 1

def trapezio(f,a,b,N):
    soma = 0
    h = abs(b-a)/N
    for i in range(1,N):
        soma += f(a + i*h)
    
    return h*(f(a)/2 + f(b)/2 +soma)

def Simpson(f,a,b,N):
    even = 0
    odd = 0
    h = abs(b-a)/N
    for i in range(1,N):
        if (-1)**i > 0:
            even += f(a+i*h)
        else:
            odd += f(a+i*h)
    return (1/3)*h*(f(a)+f(b)+4*odd+2*even)


print("Para N = 10,","trapezio: ",trapezio(f,a,b,N), "O metodo de Simpson: ", Simpson(f,a,b,N))
erroTrap = abs(Vreal - trapezio(f,a,b,N))
erroSimp = abs(Vreal - Simpson(f,a,b,N))
print("Erro trapezio: ", erroTrap, "erro Simpson: ", erroSimp)

#Agora Testar com o erro da eq 5.28 do livro
Integral1 = trapezio(f,a,b,N)
Integral2 = trapezio(f,a,b,2*N)
erroLivro528 = (1/3)* abs(Integral2 - Integral1)
print("Agora o erro da eq 5.28 do livro: ", erroLivro528)

N = 100
erroTrap = abs(Vreal - trapezio(f,a,b,N))
erroSimp = abs(Vreal - Simpson(f,a,b,N))
print("Para N = 100,","trapezio: ",trapezio(f,a,b,N), "O metodo de Simpson: ", Simpson(f,a,b,N))
print("Erro trapezio: ", erroTrap, "erro Simpson: ", erroSimp)
#Agora Testar com o erro da eq 5.28 do livro
Integral1 = trapezio(f,a,b,N)
Integral2 = trapezio(f,a,b,2*N)
erroLivro528 = (1/3)* abs(Integral2 - Integral1)
print("Agora o erro da eq 5.28 do livro: ", erroLivro528)


N = 1000
erroTrap = abs(Vreal - trapezio(f,a,b,N))
erroSimp = abs(Vreal - Simpson(f,a,b,N))
print("Para N = 1000,","trapezio: ",trapezio(f,a,b,N), "O metodo de Simpson: ", Simpson(f,a,b,N))
print("Erro trapezio: ", erroTrap, "erro Simpson: ", erroSimp)
#Agora Testar com o erro da eq 5.28 do livro
Integral1 = trapezio(f,a,b,N)
Integral2 = trapezio(f,a,b,2*N)
erroLivro528 = (1/3)* abs(Integral2 - Integral1)
print("Agora o erro da eq 5.28 do livro: ", erroLivro528)




```

```python
def Simpson(f,a,b,N):
    even = 0
    odd = 0
    h = abs(b-a)/N
    for i in range(1,N):
        if (-1)**i > 0:
            even += f(a+i*h)
        else:
            odd += f(a+i*h)
    return (1/3)*h*(f(a)+f(b)+4*odd+2*even)



def Bessel(theta,m,x):
    return np.cos(m*theta - x*np.sin(theta))

def J(m,x):
    f = Bessel(0,m,x)
    return (1/np.pi)*Simpson(f,0,np.pi,1000)

xarray = np.arange(21)
Jarray = np.zeros(21)
m = 0
for x in xarray:
    Jarray[m] = J(m,x)
    m += 1
print(Jarray)
plt.plot(Jarray,xarray)
plt.show()

```
